## FileOrganizer

```python
import os, shutil, csv


#SE NON SONO NELLA DIRECTORY "files", MI SPOSTO IN ESSA

if os.getcwd().endswith("files") != True:
    os.chdir("files")
else:
    pass


b = os.listdir()

#ESTENSIONI ACCETTATE E LISTA CARTELLE DA CREARE

acc_docs = ["txt", "odt"]

acc_images = ["png", "jpg", "jpeg"]

acc_audio = ["mp3"]

folders = ["docs", "images", "audio"]

#CREAZIONE SOTTOCARTELLE (DOCS, IMAGES, AUDIO...)

for folder in folders:
    if os.path.isdir(folder) != True:
        os.makedirs(folder)
    else:
        pass

#CREAZIONE ED HEADER FILE RECAP.CSV ("W" MODE PERCHE' DEVO SCRIVERE L'HEADER SULLA PRIMA RIGA DEL FILE)

if os.path.isfile("recap.csv") == False:
    with open('recap.csv', mode='w', newline="") as recap:
        recappino = csv.writer(recap, delimiter=',')
        recappino.writerow(["name", "type", "size(B)"])
else:
    pass

#CICLO PRINCIPALE, ITERAZIONE SUI FILES DELLA CARTELLA 'FILES' (ESCLUSO IL .CSV)
#QUI IL FILE E' APERTO IN "A" MODE PERCHE' DEVO AGGIUNGERE AL FILE, NON SOVRASCRIVERE

with open('recap.csv', mode='a', newline="") as recap:
    recappino = csv.writer(recap, delimiter=',')
    for file in sorted(b):
        if os.path.isfile(file) == True and file.split(".")[1] != "csv":
            d = os.stat(file)
            tipo = ""
            cartella = ""
            if file.split(".")[1] in acc_docs:
                tipo = "doc"
                cartella = "docs"
            elif file.split(".")[1] in acc_images:
                tipo = "image"
                cartella = "images"
            else:
                tipo = "audio"
                cartella = "audio"
            
            #STAMPO A SCHERMO LE INFORMAZIONI (NAME, TYPE, SIZE(B)) DEI FILES
            
            print(file.split(".")[0], "type:{}".format(tipo), "size:{}B".format(d.st_size))
            
            #SCRIVO LE INFORMAZIONI (NAME, TYPE, SIZE(B)) DEI FILES ALL'INTERNO DEL FILE RECAP.CSV
            
            recappino.writerow([file.split(".")[0], "{}".format(tipo), "{}".format(d.st_size)])
            
            #SPOSTO I FILES NELLE RELATIVE SOTTOCARTELLE
            
            shutil.move(file, cartella)
                
        else:
            pass
```

## Numpy (Tabella riassuntiva immagini RGBa)

```python
import os
import numpy as np
from PIL import Image
from tabulate import tabulate

#SE NON SONO IN "IMAGES" MI SPOSTO IN "IMAGES"
if os.getcwd().endswith("files") == True:
        os.chdir("images")

elif os.getcwd().endswith("images") == True:
    pass

else:
    os.chdir("files/images")



mi = os.listdir()



lista_rgb = []



for i in mi:
    
    image = Image.open('{}'.format(i)) #Apre immagine
    
    np_img = np.array(image) #Array dell'immagine

    #SE E' GRAYSCALE
    if np_img.ndim == 2:
        
        mi_grayscale = np.mean(np_img, axis = (0, 1))
        
        lista_rgb.append([i.split(".")[0], np_img.shape[0], np_img.shape[1], mi_grayscale, 0, 0, 0, 0])        
        
    
    #SE HA CANALE ALPHA
    elif np_img.shape[2] in np_img and np_img.shape[2] == 4:
        
        rosso = np_img[:,:,0]
        verde = np_img[:,:,1]
        blue = np_img[:,:,2]
        alpha_ch = np_img[:,:,3]
        
        mi_rosso = np.mean(rosso, axis = (0,1))
        mi_verde = np.mean(verde, axis = (0,1))
        mi_blue = np.mean(blue, axis = (0,1))
        mi_alpha_ch = np.mean(alpha_ch, axis = (0,1))
        
        
        lista_rgb.append([i.split(".")[0], np_img.shape[0], np_img.shape[1], 0, mi_rosso, mi_verde, mi_blue, mi_alpha_ch])
    
    
    #RGB
    else:
        
        rosso = np_img[:,:,0]
        verde = np_img[:,:,1]
        blue = np_img[:,:,2]
        
        mi_rosso = np.mean(rosso, axis = (0,1))
        mi_verde = np.mean(verde, axis = (0,1))
        mi_blue = np.mean(blue, axis = (0,1))
        
        
        lista_rgb.append([i.split(".")[0], np_img.shape[0], np_img.shape[1], 0, mi_rosso, mi_verde, mi_blue, 0])
    


print(tabulate(lista_rgb, headers = ["name", "height", "width", "grayscale", "R", "G", "B", "ALPHA"], tablefmt="fancy_grid", floatfmt = ".2f"))
```
